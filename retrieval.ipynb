```python
# BM25 Parameters
k1 = 1.5
b = 0.75
N = len(address_book)
avgdl = sum(len(doc) for doc in address_book) / N
```

```python
N, avgdl
```

```python
def compute_doc_freq(docs):
    df = {}
    for doc in docs:
        unique_terms = set(doc)
        for term in unique_terms:
            df[term] = df.get(term, 0) + 1
    return df

doc_freq = compute_doc_freq(address_book)
```

```python
def compute_idf(term, doc_freq, N):
    df = doc_freq.get(term, 0)
    return math.log((N - df + 0.5) / (df + 0.5) + 1)
```

```python
def bm25_score(query, doc, doc_index):
    score = 0.0
    doc_len = len(doc)
    term_freq = Counter(doc)
    for term in query:
        if term not in term_freq:
            continue
        idf = compute_idf(term, doc_freq, N)
        freq = term_freq[term]
        numerator = freq * (k1 + 1)
        denominator = freq + k1 * (1 - b + b * doc_len / avgdl)
        score += idf * (numerator / denominator)
    return score
```
